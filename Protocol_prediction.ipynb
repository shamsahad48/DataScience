```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('ws.csv')
```

```python
df
```

```python
dummies=pd.get_dummies(df.Source)
dummies
```

```python
dummies2=pd.get_dummies(df.Destination)
dummies2
#df.Destination.shape
```

```python
merged=pd.concat([df,dummies,dummies2],axis=1)
merged
```

```python
final = merged.drop(['Source','Destination','0.0.0.0','::1','Info'],axis='columns')
```

```python
final
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
X=final.drop(['Protocol'],axis='columns')
```

```python
Y=final.Protocol
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
```

```python
cl1 = DecisionTreeClassifier()
cl2 = svm.SVC(kernel='linear')
knn = KNeighborsClassifier(n_neighbors=5)
cl3=RandomForestClassifier(n_estimators=100)
clf = MultinomialNB()# Train Decision Tree Classifer
cl1 = cl1.fit(X_train,y_train)
cl2.fit(X_train, y_train)
knn.fit(X_train, y_train)
cl3.fit(X_train,y_train)
clf.fit(X_train, y_train)
acc=[]
```

```python
y_pred = cl1.predict(X_test)
print("Accuracy For Decision Tree:",metrics.accuracy_score(y_test, y_pred))
print("Confusion metrix for DT:\n",metrics.confusion_matrix(y_test, y_pred))
print(metrics.classification_report(y_test, y_pred, digits=3))
acc.append(metrics.accuracy_score(y_test, y_pred)*100)
#print("Sensitivity or Recall Score:",metrics.precision_score(y_test, y_pred))
```

```python
y_pred = knn.predict(X_test)
print("Accuracy For KNN classifier:",metrics.accuracy_score(y_test, y_pred))
acc.append(metrics.accuracy_score(y_test, y_pred)*100)
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
y_pred=cl3.predict(X_test)
print("Accuracy For Random Forest Classifier:",metrics.accuracy_score(y_test, y_pred))
acc.append(metrics.accuracy_score(y_test, y_pred)*100)
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
y_pred = clf.predict(X_test)
print("Accuracy For Multinominal Naive Bayes:",metrics.accuracy_score(y_test, y_pred))
acc.append(metrics.accuracy_score(y_test, y_pred)*100)
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
y_pred = cl2.predict(X_test)
print("Accuracy For Support Vector Machine:",metrics.accuracy_score(y_test, y_pred))
acc.append(metrics.accuracy_score(y_test, y_pred)*100)
print(metrics.classification_report(y_test, y_pred, digits=3))
```
